# Mark the success/failed launches for each site on the map


```python
spacex_df.columns= ['Launch_Site', 'Lat', 'Long', 'class']
spacex_df.tail(10)
```

```python
marker_cluster = MarkerCluster()
```

```python
def color_marker(Ol):
    if Ol == 1:
        return 'green'
    else:
        return 'red'
```

```python
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

```python
clusters = marker_cluster.add_to(site_map)

for lat, lng, label, col in zip(spacex_df.Lat, spacex_df.Long, spacex_df.Launch_Site, spacex_df.marker_color):
    folium.Marker(
        location=[lat, lng],
        popup=label,
        icon=folium.Icon(color='white', icon_color= col) 
        ).add_to(clusters)
site_map
```

# Calculate the distances between a launch site to its proximities


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return print(distance, 'km')
```

```python
#calculate_distance(lat1, lon1, lat2, lon2)
x = [28.57205, -80.58528]
CCAFS_rw = [28.56341, -80.57678]

distance_railway = calculate_distance(CCAFS_rw[0], CCAFS_rw[1], x[0], x[1])
distance_railway 
```

```python
markerx = folium.map.Marker(x,
                            icon=DivIcon(
                              icon_size=(20,20),
                              icon_anchor=(0,0),
                              html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % '1270 m',
                              )
                            )

site_map.add_child(markerx)
```

```python
points=(CCAFS_rw, x)
ploy = folium.PolyLine(points, color='blue')
site_map.add_child(ploy)
```
